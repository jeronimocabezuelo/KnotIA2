```python
from CustomKnot import *
from KnotDatabase import *
import pandas as pd
```

```python
def normalizeExtendedGaussCode(eGC: list[int], maxNumberOfStrands: int):
    eGC = deepcopy(eGC)
    appended = True
    while len(eGC)<maxNumberOfStrands:
        if appended:
            eGC.append(0)
        else:
            eGC.insert(0,0)
        appended = not appended
    return eGC
```

```python
masterDBImage = pd.read_csv("databases/master.csv")

xy = []
i=0
maxNumberOfStrands = 30
for index,row in masterDBImage.iterrows():
    i+=1
    codeKnot = codeKnotToNumber(row["name"])
    if codeKnot<2:
        crosses = row["crosses"]
        knot = CustomKnot(crosses)

        gaussCode = knot.extendedGaussCode
        if len (gaussCode)<=maxNumberOfStrands:
            gaussCode = normalizeExtendedGaussCode(gaussCode,maxNumberOfStrands)
            xy.append([codeKnot, gaussCode])
print(f"Con menos de {int(maxNumberOfStrands/2)} cruces hay {len(xy)} elementos de los cuales {len([t for t in xy if t[0]==0])} son de tipo 0 y {len([t for t in xy if t[0]==1])} son de tipo 1")
```
